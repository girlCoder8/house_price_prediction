```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# reading the housing data file
housing = pd.read_csv('housing.csv')

# checking the overall shape of the data
housing.shape

# printing out the top 5 rows (0-4)
housing.head()

# printing out the last 5 rows
housing.tail()

# plotting the data
housing.plot("median_income", "median_house_value")

# difficult to see individual data points; using scaatter diagrams
housing.plot.scatter("median_income", "median_house_value")

# analyzing the data and split up the data as 80% train and 20% test. The y-values will be the median_house_value, and the x-values will be the median_income.
x_train, x_test, y_train, y_test = train_test_split(housing.median_income, housing.median_house_value, test_size = 0.2)

# Imposing a linear regression
regr = LinearRegression()

# predicting; This will shape the model using one predictor. Reshape is being applied to change it from pandas to NumPy, and finally into a vector. (Reshape transverses it from a single dimension matrix to a vertical shape.)
regr.fit(np.array(x_train).reshape(-1,1), y_train)

# pass in the data to give predictions with the actual values.
preds = regr.predict(np.array(x_test).reshape(-1,1))

# let's compare our predictions
y_test.head()

# let's execute our predictions
preds

# Comparing the first values. For the actual, it is equal to 252,900. Our prediction, on the other hand, guesses approximately 180,156. (That is not bad, but that is not great!)
# Looking at values is great visually, but there are thousands of data points to be considered.
# the predicted value minus the actual test value for all the data points
residuals = preds - y_test

# plotting the histograms to see how off each value is.
plt.hist(residuals)

# using the root mean error to find the error
mean_squared_error(y_test, preds) ** 0.5

# Note: for more accuracy, you should use decision trees or random forests





```
